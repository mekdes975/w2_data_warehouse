```python
import pandas as pd

# Assuming 'df' is your DataFrame and 'column_name' is the name of the column you want to analyze
unique_counts = df['type'].value_counts()

# Display unique values and their counts
print(unique_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Plotting the bar graph
plt.figure(figsize=(10, 6))  # Set the figure size
unique_counts.plot(kind='bar')
plt.title('Frequency of type of vehicles')
plt.xlabel('types')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

          the above graph shows are frequent vehicle for the area, followed by motorcycles. 


```python
import pandas as pd

# Assuming 'df' is your DataFrame
# Group by 'type' column and aggregate using mean for specified columns
aggregated_data = df.groupby('type').agg({
    'speed': 'mean',
    'lon_acc': 'mean',
    'lat_acc': 'mean'
}).reset_index()

# Display the aggregated data
print(aggregated_data)
```

From the aggregated data extracted above, it is shown that motorcyles have the highes mean speed and heavy vehicles have the lowest. 

```python

# Assuming 'df' is your DataFrame
# Group by 'type' column and aggregate using mean for specified columns
agg_time = df.groupby('time').agg({
    'type' : 'sum'
}).reset_index()

# Display the aggregated data
print(agg_time)
```

The above aggregated data shows, how many vehicles appear at the location at that instant second of time. 

```python
# Choose the columns you want to plot ('speed' and 'traveled_d' in this case)
x_column = 'type'
y_column = 'speed'

# Plotting the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df[x_column], df[y_column], alpha=0.5)
plt.xlabel('type of vehicles')
plt.ylabel('speed')
plt.title('Scatter Plot of Speed vs Type of vehicle')
plt.grid(True)
plt.show()
```

```python
# Plotting the boxplot
plt.figure(figsize=(10, 6))
df.boxplot(column='speed', by='type', figsize=(12, 8))
plt.xlabel('Vehicle Type')
plt.ylabel('Speed')
plt.title('Boxplot of Speed by Vehicle Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.figure(figsize=(10, 6))
df.boxplot(column='lat_acc', by='type', figsize=(12, 8))
plt.xlabel('Vehicle Type')
plt.ylabel('lat_acc')
plt.title('Boxplot of latitude acceleration by Vehicle Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.figure(figsize=(10, 6))
df.boxplot(column='lon_acc', by='type', figsize=(12, 8))
plt.xlabel('Vehicle Type')
plt.ylabel('lon_acc')
plt.title('Boxplot of Speed by Vehicle Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' contains columns 'speed', 'time', and 'type'

# Get unique vehicle types
unique_types = df['type'].unique()

# Create separate scatter plots for each vehicle type
plt.figure(figsize=(10, 6))

for vehicle_type in unique_types:
    filtered_data = df[df['type'] == vehicle_type]
    plt.scatter(filtered_data['time'], filtered_data['speed'], label=vehicle_type, alpha=0.5)

plt.xlabel('Time')
plt.ylabel('Speed')
plt.title('Speed Variation over Time by Vehicle Type')
plt.xticks(rotation=45)
plt.legend(title='Vehicle Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()



```
